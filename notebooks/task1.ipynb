```python
from models.brv.square_middle_shifted_brv import ShiftedSquareMiddleBRV
from models.brv.square_middle_brv import SquareMiddleBRV
from models.brv.mul_cong_brv import MulCongBRV
from tools.frequency_test import frequency_test
from tools.r_test import r_test
import numpy as nmp

import matplotlib.pyplot as plt

plt.style.use('dark_background')
```

```python

k = 10

def plot(data : nmp.ndarray, label : str):
    x = [f'{i + 1}' for i in range(k)]
    fig, axs = plt.subplots(1, 2, sharey='row', tight_layout=True)
    axs[0].bar(x, data[0], label='n=100')
    axs[0].legend()
    axs[1].bar(x, data[1], label='n=10000')
    axs[1].legend()
    fig.suptitle(label)
    plt.show()
```

```python

sm_frequency = frequency_test(
    SquareMiddleBRV(),
    k=k
)
plot(sm_frequency , 'SquareMiddleBRV')


mc_frequency = frequency_test(
    MulCongBRV(),
    k=k
)
plot(mc_frequency, 'MulCongBRV')

ssm_frequency = frequency_test(
    ShiftedSquareMiddleBRV(),
    k=k
)
plot(ssm_frequency, 'ShiftedSquareMiddleBRV')



```

```python

x = (nmp.arange(100) + 1) * 100
sm_r = r_test(SquareMiddleBRV(), x)
ssm_r = r_test(ShiftedSquareMiddleBRV() , x)
mc_r = r_test(MulCongBRV(), x)

plt.plot(x, sm_r, label='SquareMid')
plt.plot(x, mc_r, label='MulCong')
plt.plot(x, ssm_r, label='ShiftedSquareMid')
plt.legend()
plt.show()

```
